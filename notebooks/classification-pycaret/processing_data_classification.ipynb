### Criação de uma nova coluna 'data_voo' por meio das colunas de data

```python
# Criando a coluna de data voo
parquet['data_voo'] = 0 
```

```python
# Definição da função que cria a data do voo a partir de uma linha (row) do Dataframe
def create_data_voo(row):
    # Extrai o ano, mês e dia das colunas do DataFrame
    year = int(row['dateYear-1'])
    month = int(row['dateMonth-1'])
    day = int(row['dateDay-1'])

    # Verifica se a data (mês e dia) é válida
    if 1 <= month <= 12 and 1 <= day <= 31:
        return f"{year}-{str(month).zfill(2)}-{str(day).zfill(2)}"
    else:
        return pd.NaT  # Retorna "Not a Timestamp" para datas inválidas

# Aplica a função 'create_data_voo' em cada linha do DataFrame para criar a coluna 'data_voo'
parquet['data_voo'] = parquet.apply(create_data_voo, axis=1)
parquet['data_voo'] = pd.to_datetime(parquet['data_voo'], errors='coerce')
```

```python
df= parquet
```

```python
# Dropando as primeiras colunas
df = parquet.drop([0, 1, 2])
df.reset_index(drop=True, inplace=True)
df
```

```python
# Preenchendo a coluna de data vazia 
df['data_voo'] = pd.to_datetime(df['data_voo'])
last_valid_datetime = None


# Neste loop vamos preenchendo todas as colunas de data com as últimas datas 
for index, row in df.iterrows():
    if pd.isnull(row['data_voo']):
        # Checa se é NaN e coloca a última data válida
        df.at[index, 'data_voo'] = last_valid_datetime
    else:
        # Se não for NaN atualiza a data válida
        last_valid_datetime = row['data_voo']
```

### Estabelecimento de uma janela semanal para identificação de falhas. Se for detectada uma falha em qualquer dia específico, o sistema retroagirá 6 dias, englobando assim um intervalo total de 7 dias e classificando-o como um intervalo de falha. Para essa identiticação será adicionada uma coluna chamada 'failure_week' ao conjunto de dados. Esta coluna indicará se houve ou não uma falha durante aquele intervalo semanal específico, por meio de decodificação binária de 0 e 1. A identificação de falha será feita por meio das colunas de 'message'.




```python
# Criando a flag de falhas semanal 
df_data['failure_week'] = 0
```

```python
# Definindo as datas para criar o intervalo de falhas semanais
begin_fail_1 = pd.to_datetime('2022-05-23')  
finish_fail_1 = pd.to_datetime('2022-05-30')

begin_fail_2 = pd.to_datetime('2022-05-25')  
finish_fail_2 = pd.to_datetime('2022-06-01')

begin_fail_3 = pd.to_datetime('2022-05-27')  
finish_fail_3 = pd.to_datetime('2022-06-03')

begin_fail_4 = pd.to_datetime('2022-10-30')  
finish_fail_4 = pd.to_datetime('2022-11-06')

begin_fail_5 = pd.to_datetime('2022-12-27')  
finish_fail_5 = pd.to_datetime('2023-01-03')

begin_fail_6 = pd.to_datetime('2022-10-30')  
finish_fail_6 = pd.to_datetime('2022-11-06')

begin_fail_7 = pd.to_datetime('2022-07-07')  
finish_fail_7 = pd.to_datetime('2022-07-14')
```

```python
# Intervalos de falha
date_intervals = [
    (pd.to_datetime('2022-05-23'), pd.to_datetime('2022-05-30')),
    (pd.to_datetime('2022-05-25'), pd.to_datetime('2022-06-01')),
    (pd.to_datetime('2022-05-27'), pd.to_datetime('2022-06-03')),
    (pd.to_datetime('2022-10-30'), pd.to_datetime('2022-11-06')),
    (pd.to_datetime('2022-12-27'), pd.to_datetime('2023-01-03')),
    (pd.to_datetime('2022-10-30'), pd.to_datetime('2022-11-06')),
    (pd.to_datetime('2022-07-07'), pd.to_datetime('2022-07-14'))
]
```

```python
# Converte a coluna 'data_voo' em TimeStamp
df_data['data_voo'] = pd.to_datetime(df_data['data_voo'])
```

```python
# Cria uma nova coluna 'failure_week' que inicia com 0
df_data['failure_week'] = 0
```

```python
# Defini 'failure_week' como 1 para linhas correspondentes
for start_datetime, end_datetime in date_intervals:
    conditional = (df_data['data_voo'] >= start_datetime) & (df_data['data_voo'] <= end_datetime)
    df_data.loc[conditional, 'failure_week'] = 1
```
