```python
from pycaret.regression import *
import pandas as pd
```

### Leitura do DataFrame

```python
df = pd.read_csv('/content/dataset_processado_06120018_more_intervals.csv')
```

```python
# tira a coluna de index que aparece ao importar o arquivo
df = df.drop(columns={'Unnamed: 0'})
```

## Treinamento do modelo com Pycaret

```python
# seta as configurações do autoML
s = setup(df, target = 'dias_faltando_coluna',
          session_id = 123,
          normalize=True,
          use_gpu=True)
```

```python
# compara os melhores modelos
best = compare_models()
```

```python
# salva a pipeline
save_model(best[0], 'extra_trees')
```

```python
evaluate_model(best[0])
```

```python
# cross validation

s = setup(df, target = 'dias_faltando_coluna',
          session_id = 123,
          normalize=True,
          fold = 5,
          use_gpu=True)
```

```python
cross = compare_models()
```

```python
megamodelotron_tunado_3000_no_virus_atualizado_2023 = tune_model(cross[0])
```
