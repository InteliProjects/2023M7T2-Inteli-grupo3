```python
import pandas as pd
import dask.dataframe as dd
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
from pycaret.regression import *
```

```python
df = df.ffill()
df = df.bfill()

df.head()
```

```python
df.head()
```

```python
df['message'] = df['message0418DAA-1'] + df['message0422DAA-1']

df.head()
```

```python
colunas_remover = ['message0418DAA-1', 'message0422DAA-1']
df = df.drop(columns=colunas_remover)
df.head()
```

```python
s = setup(
    data=df,
    use_gpu=True,
    target = "message",
    session_id = 152
)
```

```python
melhor_modelo = compare_models()
```

```python
melhor_modelo
```

```python
evaluate_model(melhor_modelo)
```

```python
save_model(melhor_modelo, 'modeloETRpycaret')
```
